Importing Libraries and Datasets


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df = pd.read_csv("/content/drive/My Drive/Garbage classification/data.csv",encoding = "ISO-8859-1")
df.head()
```

Cleaning the Data

```
df=df.drop(['agency','stn_code','location_monitoring_station','sampling_date'], axis = 1)
```

```
df.head()
```

```
replacements = {
    'state': {
        'Uttaranchal' : 'Uttarakhand'
    }
}
df.replace(replacements,regex=True,inplace=True)
```

```
df['date'] = pd.to_datetime(df['date'])
```

```
df['year'] = df['date'].dt.year
```

```
df['type'].value_counts().plot.bar()
```

Distribution and Trend of Nitrogen Dioxide(No2) emission

```
df_group_by_no2 = df.groupby(['year']).median().reset_index().sort_values(by="year", ascending="False")
```

```
f,ax=plt.subplots(figsize=(15,5))
plt.xticks(rotation=90)
sns.pointplot(x="year", y="no2", data=df_group_by_no2)
```

```
df_group_by_no2_1 = df.groupby([df['type']])['no2'].median().sort_values(ascending=False)
df_group_by_no2_1.plot.bar(color='y')
```

Distribution and Trend of Sulphur Dioxide(So2) emission

```
f,ax = plt.subplots(figsize=(15,5))
plt.xticks(rotation=90)
sns.pointplot(x='year',y='so2',data=df_group_by_no2)
```

```
df_group_by_so2_1 = df.groupby([df['type']])['so2'].median().sort_values(ascending=False)
df_group_by_so2_1.plot.bar(color='r')
```

Distribution of No2 (based on month)


```
df_group_by_month_no2 = df.groupby([df['date'].dt.strftime('%B')])['no2'].mean().sort_values(ascending=False)
df_group_by_month_no2.plot.bar(color='r')
```

Correlation Between Features

```
df['pm2_5'].corr(df['so2'])
```

Takeaway: There is no much correlation 2.5ppm and s02 as it is less than 0.5

```
df['pm2_5'].corr(df['no2'])
```

Takeway: We can see that no2 is more correlated to 2.5ppm than so2.So If we reduce no2 we can reduce the dangerous 2.5 ppm

```
df['rspm'].corr(df['pm2_5'])
```

Takeaway: Rspm is much correlated to 2.5ppm

Distribution of Sulphur dioxide(So2) based on cities

```
df_group_by_state_so2 = df.groupby([df['state']])['so2'].median().sort_values(ascending=False)
df_group_by_state_so2.plot.bar(color='r')
```

Distribution of No2 emission based on states

```
df_group_by_state_no2 = df.groupby([df['state']])['no2'].median().sort_values(ascending=False)
df_group_by_state_no2.plot.bar(color='g')
```

Takeaway: Uttarkhand has the highest amount of So2 emission and West Bengal has the highest amount of No2 emission. States like Jharkhand,Bihar,Maharastra and Sikkim is among the top states in both the list

<h1>Comparison between the top states of both the list(West Bengal and Uttarkhand)</h1>

<h4>No2 emissions</h4>

```
df_wb_group_year = df[df['state']=='West Bengal'].groupby('year').median().reset_index().sort_values(by=['no2'],ascending=False)
```

```
df_ut_group_year = df[df['state']=='Uttarakhand'].groupby('year').median().reset_index().sort_values(by=['no2'],ascending=False)
```

```
f,ax=plt.subplots(figsize=(15,5))
plt.xticks(rotation=90)
sns.pointplot(x='year', y='no2', data=df_wb_group_year,ax=ax,color='g');
sns.pointplot(x='year', y='no2', data=df_ut_group_year,ax=ax);
```

Takeaway: While seeing this we can obtain that the rates in West Bengal is always higher than that of Uttarkhand.But during 2015 the rates has been good

<h4>So2 emission</h4>

```
df_wb_group_years = df[df['state']=='West Bengal'].groupby('year').median().reset_index().sort_values(by=['so2'],ascending=False)
df_ut_group_years = df[df['state']=='Uttarakhand'].groupby('year').median().reset_index().sort_values(by=['so2'],ascending=False)
```

```
f,ax=plt.subplots(figsize=(15,5))
plt.xticks(rotation=90)
sns.pointplot(x='year', y='so2', data=df_wb_group_years,ax=ax,color='g');
sns.pointplot(x='year', y='so2', data=df_ut_group_years,ax=ax);
```

Takeaway: Although West Bengal has higher rates than Uttarkhand but Uttsrkhand is rated at top is because of the dip in the rates in west bengal durig 2015 .The dip could have been some stratigies or awareness done by the government or some discripncy in data
